## Read Data from Silver Layer
Let's read the parquet files that we created in the Silver layer!

Don't worry, it's the same as the output from your exercise (if all of your tests passed)!

```
from pyspark.sql import DataFrame
df = spark.read.parquet("dbfs:/FileStore/vinod.murugesan/heka_batch_processing_silver/output_silver/")
print(df.count())
display(df)
```

```
import pyspark.sql.functions as fn
from pyspark.sql.types import IntegerType
df_MaxProcess=df.withColumn("Process_total",fn.col("Process_total").cast(IntegerType())) \
.select('Category','Process_total').groupBy('Category').agg(fn.max('Process_total').alias('Total_Processes_Breached'))
display(df_MaxProcess)
```
